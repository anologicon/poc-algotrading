```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
```

```python
dfTopAssets = pd.read_csv('stock_assets_100.csv')

tickers = dfTopAssets['Symbol']+'.SA'
```

```python
portifolio = pd.DataFrame();
for t in tickers:
    portifolio[t] = wb.DataReader(t, data_source='yahoo', start='2021-2-1')['Adj Close']
```

```python
portifolio.corr()
```

```python
pd.concat([portifolio, portifolio.shift(), portifolio.shift(2)], axis=1)
```
